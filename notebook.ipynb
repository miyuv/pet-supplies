```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#We load and print our dataset.
pet_data = pd.read_csv('pet_supplies_2212.csv')
print(pet_data.head())
```

```python
#What is the datatype of #product_id?
print(pet_data['product_id'].dtype)
```

```python
#How many product_id values are there?
tr_productid = pet_data['product_id'].count()
print(f"Total number of data in column product_id : {tr_productid}")

# Check unique value count for product_id
uc_productid = pet_data['product_id'].nunique()
print(f"Number of unique data in column product_id: {uc_productid}")

#All values are unique because they are equal.
```

```python
# Check for missing values for product_id
mv_product = pet_data['product_id'].isnull().sum()
print(mv_product)

# No missing values.
# No changes have been made to this column.
```

```python
#What is the datatype of #category?
print(pet_data['category'].dtype)
```

```python
#check for missing values for category
mv_category = pet_data['category'].isnull().sum()
print(mv_category)

#No missing values.
```

```python
#Print unique values in #category.
print(pet_data['category'].unique())

#There is an undesirable "-" value, we should change it to Unknow.
pet_data['category'] = pet_data['category'].replace('-', 'Unknown')

#Let's check it out.
print(pet_data['category'].unique())

#category had data "-". I changed it to Unknow. Other than that, there is no missing data.
```

```python
#What is animal's data type?
print(pet_data['animal'].dtype)
```

```python
# check for missing values for animal
mv_animal = pet_data['animal'].isnull().sum()
print(mv_animal)

#No missing values.
```

```python
#Print unique values in animal.
print(pet_data['animal'].unique())

#Unique values as said. No missing values. The data is nominal. No changes were made.
```

```python
#What is size's data type?
print(pet_data['size'].dtype)
```

```python
#  check for missing values for size
mv_size = pet_data['size'].isnull().sum()
print(mv_size)

#No missing values.
```

```python
#Print the unique values in size.
print(pet_data['size'].unique())

# convert the values in the size column to lowercase 
pet_data['size'] = pet_data['size'].str.lower() 

# Let's fix the data. 
pet_data['size'].replace(['small', 'medium', 'large'], ['Small', 'Medium', 'Large'], inplace=True)
```

```python
#What is the datatype of price?
print(pet_data['price'].dtype)

#We need to convert the data type from object to float. We could not convert the data type to float because it has #Unlisted value. First we have to write 0 instead of unlisted.

pet_data['price'] = pet_data['price'].replace('unlisted', 0)

# Now we can convert it to float.
pet_data['price'] = pet_data['price'].astype(float)

# Let's see if it happened.
print(pet_data['price'].dtype)
```

```python
# replace the 0 values in the "price" column with the overall average price
avg_price = pet_data['price'].mean()
pet_data['price'] = pet_data['price'].replace(0, avg_price)
```

```python
# check for missing values for price
mv_price = pet_data['price'].isnull().sum()
print(mv_price)

# No missing values.
```

```python
# check the presence of negative values for price
negatif_var_mi_price = (pet_data['price'] < 0).any()

if negatif_var_mi_price:
    print("Negative Values Found.")
else:
    print("No Negative Value Found.")
    
#No negative value.

#The data type of the #Price column was object, we needed to convert it to float, but we got an error because the column had an 'unlisted' value.
#So first unlisted values are replaced with 0. The data type was converted to float, then we replaced 0 with the overall average.
#Other than that, there were no missing, negative, or unwanted values.
```

```python
#What is the sales data type?
print(pet_data['sales'].dtype)
```

```python
# sales check for negative values
negatif_var_mi_sales = (pet_data['sales'] < 0).any()

if negatif_var_mi_sales:
    print("Negative Values Found.")
else:
    print("No Negative Value Found.")
    
#No negative value.
```

```python
#sales missing value check
mv_sales = pet_data['sales'].isnull().sum()
print(mv_sales)
# No missing values.

# There is no other situation, no change has been made.
```

```python
#What is the rating data type?
print(pet_data['rating'].dtype)
```

```python
#rating missing value check
mv_rating = pet_data['rating'].isnull().sum()
print(mv_rating)

# There is a missing value. We have to replace it with 0.
pet_data['rating'].fillna(0, inplace=True)

# Let's print it out.
print(mv_rating)
```

```python
#Print unique values in rating.
print(pet_data['rating'].unique())
```

```python
#Missing values in repeat_purchase?
mv_repeat_purchase=pet_data['repeat_purchase'].isnull().sum()
print(mv_repeat_purchase)

# No missing values.
```

```python
#Unique values in repeat_purchase?
print(pet_data['repeat_purchase'].unique())

#There were two unique values that matches the two given in the data dictionary. There were no missing values so no changes were made to this column.
```

## Task 2

```python
df = pd.DataFrame(pet_data)

# Create a column that counts repeat purchases by category
df['repeated_purchases'] = df.groupby('category')['repeat_purchase'].transform('sum')

main_color = 'skyblue'

# Draw a bar chart using the data frame
plt.figure(figsize=(10, 6)) 
sb.barplot(data=df, x='repeated_purchases', y='category', color=main_color)
plt.title("Distribution of Categories of Repeat Purchases")
plt.xlabel("Number of Purchases")
plt.ylabel("Categories")
```

The most repeated purchase was in the equipment category.
There are 7 categories in the data. These are: Equipment, Housing, Medicine, Food, Accessories, Toys and Unknown. The highest repeat purchase category is equipment. Housing, Pharmaceuticals, Food, Toys have close repeat purchases. Out of these 7 categories, only 2 categories, accessories and unknowns, are missing from the median and average of repeat sales. The remaining 5 categories are higher than the median and average of repeat sales. Efforts should be made to increase repeat purchases in the Accessory and Unknown category.
